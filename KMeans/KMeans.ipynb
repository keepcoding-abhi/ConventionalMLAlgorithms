```python
from math import sqrt
def find_euclidean_distance (row1, row2) :
    euclidean_distance = 0
    for column_index in range(len(dataset.columns)) :
        first = row1[column_index]
        second = row2[column_index]
        if type(first) == str :
            if first != second :
                euclidean_distance += 1
        else :
            euclidean_distance += ((first - second) ** 2)
    return sqrt(euclidean_distance)
```

```python
def find_closest_cluster(current_tuple, centroids) :
    distance = None
    closest_cluster = -1
    for cluster in centroids :
        current_centroid = centroids[cluster]
        current_distance = find_euclidean_distance(current_tuple, current_centroid)
        if distance == None :
            distance = current_distance
            closest_cluster = cluster
        elif distance > current_distance :
            distance = current_distance
            closest_cluster = cluster
    return closest_cluster
```
