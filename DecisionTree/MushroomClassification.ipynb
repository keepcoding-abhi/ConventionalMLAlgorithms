```python
input_train, input_test, output_train, output_test = train_test_split(input_variables, output_variable, test_size = 0.25)
```

```python
def calculate_base_entropy(X, Y) :
    to_predict = Y.keys()[0]
    entropy = 0
    values = Y[to_predict].unique()
    for value in values :
        fraction = Y[to_predict].value_counts()[value]/len(Y)
        entropy += (-(fraction * np.log2(fraction)))
    return entropy
```

```python
eps = np.finfo(float).eps
def calculate_conditional_entropy(X, Y, attribute) :
    to_predict = Y.keys()[0]
    possible_values = Y[to_predict].unique()
    attribute_values = X[attribute].unique()
    conditional_entropy = 0
    for attribute_value in attribute_values :
        entropy = 0
        for outcome_value in possible_values :
            attribute_booleans = (X[attribute] == attribute_value)
            target_booleans = (Y[to_predict] == outcome_value)
            numerator = len(Y[attribute_booleans & target_booleans])
            denominator = len(X[attribute_booleans])
            fraction = numerator / (denominator + eps)
            entropy += (-(fraction * np.log2(fraction + eps)))
        weight = denominator / len(X)
        conditional_entropy += (weight * entropy)
    #print("Conditional entropy for attribute {} is {}.".format(attribute, conditional_entropy))
    return abs(conditional_entropy)
```

```python
def find_most_informative_attribute(X, Y) :
    information_gains = []
    base_entropy = calculate_base_entropy(X, Y)
    #print("Base entropy {}.".format(base_entropy))
    for attribute in X.keys() :
        information_gains.append(base_entropy - calculate_conditional_entropy(X, Y, attribute))
    #print("Information gains :")
    #print(information_gains)
    return X.keys()[information_gains.index(max(information_gains))]
```

```python
def make_partition(X, Y, attribute, value) :
    x_partition = X[X[attribute] == value].reset_index(drop = True)
    y_partition = Y[X[attribute] == value].reset_index(drop = True)
    return x_partition, y_partition
```

```python
def buildTree(X, Y):
    
    tree = None
    output_variable = Y.keys()[0]
    splitting_attribute = find_most_informative_attribute(X, Y)
    test_outcomes = X[splitting_attribute].unique()
    
    #print("Splitting Attribute {}.".format(splitting_attribute))
    if tree is None :
        tree = {}
        tree[splitting_attribute] = {}
    
    for outcome in test_outcomes :
        
        x_partition, y_partition = make_partition(X, Y, splitting_attribute, outcome)
        #print("Length of x partition{}, y partition {}".format(len(x_partition), len(y_partition)))
        output_values_after_partition = y_partition[output_variable].unique()
        #print("Output values after partition :")
        #print(output_values_after_partition)
        if len(output_values_after_partition) == 1 :
            tree[splitting_attribute][outcome] = output_values_after_partition[0]
        else :
            tree[splitting_attribute][outcome]  = buildTree(x_partition, y_partition)
    
    return tree
```

```python
tree1 = buildTree(input_train, output_train)
```

```python
def predict(test_tuple, tree) :
    for attribute in tree.keys() :
        value = test_tuple[attribute]
        tree = tree[attribute][value]
        prediction = 0
        
        if type(tree) is dict :
            prediction = predict(test_tuple, tree)
        else :
            prediction = tree
            break;
    return prediction
```
